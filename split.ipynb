```python
import json
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# 讀取處理過的 JSON 檔案
with open('processed_output.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 轉換為 DataFrame
df = pd.DataFrame.from_dict(data, orient='index')

# 顯示資料基本資訊
print("資料集大小:", df.shape)
print("\n前五筆資料:")
display(df.head())

# 首先分割出測試集 (20%)
train_val_data, test_data = train_test_split(
    df, 
    test_size=0.2, 
    random_state=42
)

# 再將剩下的資料分割為訓練集 (64%) 和驗證集 (16%)
train_data, val_data = train_test_split(
    train_val_data, 
    test_size=0.2, 
    random_state=42
)

# 顯示分割後的資料集大小
print("\n分割後的資料集大小:")
print(f"訓練集: {train_data.shape} ({len(train_data)/len(df)*100:.1f}%)")
print(f"驗證集: {val_data.shape} ({len(val_data)/len(df)*100:.1f}%)")
print(f"測試集: {test_data.shape} ({len(test_data)/len(df)*100:.1f}%)")

# 將分割後的資料集儲存為 JSON 檔案
def save_to_json(data, filename):
    data_dict = data.to_dict(orient='index')
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data_dict, f, ensure_ascii=False, indent=4)

# 儲存分割後的資料集
save_to_json(train_data, 'train_data.json')
save_to_json(val_data, 'val_data.json')
save_to_json(test_data, 'test_data.json')

# 顯示每個資料集的基本統計資訊
print("\n訓練集的基本統計資訊:")
display(train_data.describe())

print("\n驗證集的基本統計資訊:")
display(val_data.describe())

print("\n測試集的基本統計資訊:")
display(test_data.describe())

# 檢查特徵分布是否平衡
numeric_features = train_data.select_dtypes(include=[np.number]).columns

import matplotlib.pyplot as plt
import seaborn as sns

# 繪製主要數值特徵的分布圖
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numeric_features[:9], 1):  # 只顯示前9個特徵
    plt.subplot(3, 3, i)
    sns.kdeplot(data=train_data[feature], label='Train', alpha=0.5)
    sns.kdeplot(data=val_data[feature], label='Val', alpha=0.5)
    sns.kdeplot(data=test_data[feature], label='Test', alpha=0.5)
    plt.title(feature)
    plt.legend()
plt.tight_layout()
plt.show()
```
